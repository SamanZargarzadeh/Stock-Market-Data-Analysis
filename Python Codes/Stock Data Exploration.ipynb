## Merging all the dataset and adding returns 

```python
full=pd.DataFrame()
i=0

for t in TickInd.Symbol:
    try:
        df = yf.download(t, start="2015-01-01", end="2022-10-30")
        df["Ticker"]=t
        df["Sector"]=TickInd.Sector.iloc[i]
        
        #Growth one dollar
        first=df.Close.iloc[0]
        inv=1 #one dollar
        vol=inv/first
        df["Growth"]=df["Close"]*vol
        
        df["Monthly Returns"]=""
        df["Yearly Returns"]=""
        
        #Performance
        for k in range(30,len(df)):
            df["Monthly Returns"].iloc[k]=((df["Growth"].iloc[k]-df["Growth"].iloc[k-30])/ (df["Growth"].iloc[k-30]))*100
        
        for j in range(365,len(df)):
            df["Yearly Returns"].iloc[j]=((df["Growth"].iloc[j]-df["Growth"].iloc[j-365])/ (df["Growth"].iloc[j-365]))*100
        

        
        full=pd.concat([full, df], axis=0)
        i=i+1
    except:
        i=i+1
          
```

# Recommendation Explore

This part is ignored in the report and slides because of the limitions of time and pages. but in this part we use recommenders signal and try to compare returns we can have based on the recommenders.

```python
dfREC=pd.DataFrame()
i=0
TickInd.Symbol
for t in TickInd.Symbol:
    try:

        print(t)
        rec = yf.Ticker(t).recommendations
        rec['YMD'] = rec.index.date

        hist = yf.download(t, start=rec.YMD.iloc[0], end=rec.YMD.iloc[-1])

        hist['YMD'] = hist.index.date
        hist
        
        full = pd.merge(rec,hist, how="outer" , on='YMD')
        full=full.sort_values(by=['YMD'])
        full=full.set_index(["YMD"])
        full.head()
        
        full["position"]=np.where(full["To Grade"] == 'Buy', 1 , np.nan)
        full["position"]=np.where(full["To Grade"] == 'Long-Term Buy', 1 , full["position"])
        full["position"]=np.where(full["To Grade"] == 'Underperform', 1 , full["position"])
        full["position"]=np.where(full["To Grade"] == 'Underweight', 1 , full["position"])
        full["position"]=np.where(full["To Grade"] == 'Strong Buy', 1 , full["position"])


        full["position"]=np.where(full["To Grade"] == 'Overweight' , -1 , full["position"])
        full["position"]=np.where(full["To Grade"] == 'Sell' , -1 , full["position"])
        full["position"]=np.where(full["To Grade"] == 'Outperform' , -1 , full["position"])
        
        
        full["FULLposition"]=full["position"].ffill ()
        
        df=full
        
        df["Ticker"]=t
        df["OneREC"]=1
        s=1/df["Close"].iloc[0]

        for i in range(1,len(df)):
            s=(df["OneREC"].iloc[(i-1)])/df["Close"].iloc[(i-1)]
    
            if df["FULLposition"].iloc[i]==1:
                df["OneREC"].iloc[i]=s*df["Close"].iloc[i]

            else:
                df["OneREC"].iloc[i]=df["OneREC"].iloc[(i-1)]
        share=1/df["Close"].iloc[0]
        share
        #Buy / Hold
        df["Share"]=share
        df["One"]=df["Share"]*df["Close"]
        
        df["holdReturns"]=(df['One'].dropna().iloc[-1]-df['One'].dropna().iloc[0])*100
        df["RecReturns"]=(df['OneREC'].dropna().iloc[-1]-df['OneREC'].dropna().iloc[0])*100
        
        df_returns=df[["Ticker","holdReturns","RecReturns"]].iloc[0]

        
        dfREC=pd.concat([dfREC, df_returns], axis=1)
        i=i+1
    except:
        i=i+1
          
```

```python
reccap = dfREC.merge(Mar, how='inner', left_on='Ticker', right_on='index')
RecCapSec = reccap.merge(TickInd, how='inner', left_on='Ticker', right_on='Symbol')
#RecCapSec.to_csv('/Users/macbook/Desktop/Rec.csv', index=False, header=True)
```

HoldWin would be the tickers that buy and hold strategy give us better returns than trade based on the recommender signals and the RecWin is the time that Recommender's strategy have more returns

```python
HoldWin=RecCapSec[RecCapSec["RecReturns"]<RecCapSec["holdReturns"]]
```

```python
RecWin=RecCapSec[RecCapSec["RecReturns"]>RecCapSec["holdReturns"]]
```

```python
HoldCount=HoldWin.groupby("Sector").count().reset_index()
hc=HoldCount[["Sector","Symbol"]]
hc
###
RecCount=RecWin.groupby("Sector").count().reset_index()
rc=RecCount[["Sector","Symbol"]]
rc
####
TotCount=RecCapSec.groupby("Sector").count().reset_index()
tc=TotCount[["Sector","Symbol"]]
tc

###
hr=pd.merge(hc,rc, on="Sector")
tot=pd.merge(hr,tc, on="Sector")
tot["Rec%"]=tot["Symbol_y"]/tot["Symbol"]*100
tot["Hold%"]=tot["Symbol_x"]/tot["Symbol"]*100
tot
```

conclusion: in general buy-and-hold is better than Yfinance recommenders' strategies and it is not dependent on the sectors
